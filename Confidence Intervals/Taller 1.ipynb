**IC para Media**

image.png

```python
#sigmaP conocida
from scipy.stats import norm
n=13
Xme=52.23
Desc=14.8557
Alpha=0.05/2
Z=norm.ppf(0.95+Alpha)

print(Xme-Z*(Desc/(n**0.5)),' / Xmed / ',Xme+Z*(Desc/(n**0.5)))
```

```python
#sigmaP NO conocida
from scipy.stats import t
n=13
Xme=52.23
Des=14.8557
Alpha=0.05/2
T=t.ppf(1-Alpha,n-1)

print(Xme-T*(Des/(n**0.5)),' / Xmed / ',Xme+T*(Des/(n**0.5)))
```

**IC para diferencia de Medias**

image.png

```python
#sigmaP conocida
from scipy.stats import norm
n1=45
n2=35
Xme1=548
Xme2=517
Desc1=57
Desc2=52
Alpha=0.05/2
Z=norm.ppf(0.95+Alpha)

print((Xme1-Xme2)-Z*((((Desc1**2)/n1)+((Desc2**2)/n2))**0.5),' / Xmed1-Xmed2 / ',(Xme1-Xme2)+Z*((((Desc1**2)/n1)+((Desc2**2)/n2))**0.5))
```

```python
#sigmaP NO conocida
#sigma1 deferente sigma2

from scipy.stats import t
n1=45
n2=35
Xme1=548
Xme2=517
Des1=57
Des2=52
Alpha=0.05/2
v=((((Des1**2)/n1)+((Des2**2)/n2))**2)/(((((Des1**2)/n1)**2)/(n1-1))+((((Des2**2)/n2)**2)/(n2-1)))
T=t.ppf(1-Alpha,v)

print((Xme1-Xme2)-T*((((Des1**2)/n1)+((Des2**2)/n2))**0.5),' / Xmed1-Xmed2 / ',(Xme1-Xme2)+T*((((Des1**2)/n1)+((Des2**2)/n2))**0.5))
```

```python
#sigmaP NO conocida
#sigma1 igual sigma2

from scipy.stats import t
n1=45
n2=35
Xme1=548
Xme2=517
Des1=57
Des2=52
Desp=(((n1-1)*Des1+(n2-1)*Des2)/(n1+n2-2))**0.5
Alpha=0.05/2
T=t.ppf(1-Alpha,n1+n2-2)

print((Xme1-Xme2)-T*Desp*(((1/n1)+(1/n2))**0.5),' / Xmed1-Xmed2 / ',(Xme1-Xme2)+T*Desp*(((1/n1)+(1/n2))**0.5))
```

**Distribucion Media Muestral**

```python
#varias variables

Med=1.35
Sig=0.35
n=5

print(norm.cdf(8,n*Med,(n*(Sig**2))**0.5)-norm.cdf(6,n*Med,(n*(Sig**2))**0.5))

Z1=(8-n*Med)/((n**0.5)*Sig)
Z2=(6-n*Med)/((n**0.5)*Sig)
norm.cdf(Z1)-norm.cdf(Z2)
```

```python
#media muestral

Med=1.35
Sig=0.35
n=5

print(norm.cdf(2,Med,((Sig**2)/n)**0.5))

#o tambien

Z=(2-Med)/(Sig/(n**0.5))
norm.cdf(Z)
```
